```python
def calculate_final_price(price, discount_rate) :
    price -= price * discount_rate/100
    return price
print ("for this product\nthe price of product is 20150\nand on tis product you got 7.95% discount\nand your final price is : ",calculate_final_price(20150,7.95))
```

# question 3

```python
def format_price(price, currency='INR'):
    if currency.lower()=="inr":
        return f"₹{price}"
    elif currency.lower()=="usd":
        return f"${price}"
print(format_price(500))
print(format_price(500,"USD"))
```

```python
def apply_coupon(price, coupon_code=None):
    if coupon_code=='ZOMATO10':
        price -= price/10
        return price 
    else :
        return price
print ("your price is 560 and you applyed coupon code to 'ZOMATO10' \n so your final price is :",apply_coupon(560,"ZOMATO10"))
```
