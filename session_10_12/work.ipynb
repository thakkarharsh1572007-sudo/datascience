# question 3

```python
from functools import reduce
cart= [499, 1299, 299, 799]
print ("total cart amount is :",reduce(lambda a,b: a+b ,cart))
```

# question 4

```python
follower_counts = [950, 1500, 25000, 1200000]
ans = list (map (lambda a: f'{a/1000000.0}M' if a>=1000000 else (f'{a/1000.0}K' if a>=1000 else a),follower_counts))
print ("final output is ",ans)
```

```python
scores=[101, 98, 120, 77, 88]
even =list(filter(lambda a: a%2==0 ,scores))
odd =list(filter(lambda a: a%2==1 ,scores))
print (f"even score = {even} and odd score = {odd}")
```
