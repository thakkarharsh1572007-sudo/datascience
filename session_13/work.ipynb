# question 2

```python
def count_unread_messages(messages) :
    if not messages:
        return 0
    else :
        un = messages.popitem()[1]+ count_unread_messages(messages)
        return un
    
message = {
    "group 1" : 20,
    "group 2" : 2,
    "group 3/subgroup 1" : 10,
    "group 3/subgroup 2" : 3,
    "group 4" : 0
}
print("total underated massages are :",count_unread_messages(message))
```

# question 3

```python
x = 'global' # this is a global variable

def outer():
    x = 'outer' # this is a local variable to outer()
    
    def inner():
        nonlocal x
        x = 'inner' # this modifies outer()'s local variable, NOT the global one
        
    inner()
    print('Inside outer:', x)

outer()
print('Outside:', x)
```

```python
# def formate (n) : 
#     if n >= 1000000:
#         return f"{n/1000000.0}M"
#     elif n> 1000:
#         return f"{n/1000.0}K"
#     else :
#         return str(n)
form = lambda n : f"{n/1000000.0}M" if n>=1000000 else f"{n/1000.0}K" if n>1000 else str(n)
def format_number_short(n, r = None):
    if r == None :
        r=[]
    if not n:
        return r
    r.insert(0,form(n.pop()))
    format_number_short(n,r)
    return r
n =[10000,10000000,2026,3505]
r = format_number_short(n)
print(r)
```
